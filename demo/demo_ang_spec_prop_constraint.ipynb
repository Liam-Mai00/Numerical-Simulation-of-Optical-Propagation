# Fresnel diffraction integral sampling constraints

There are 4 inequalities that govern the sampling required for Fresnel diffraction.

1. $\delta_2 \le -\frac{D_2}{D_1}\delta_1 + \frac{\lambda \Delta z}{D_1}$
1. $N \ge \frac{D_1}{2\delta_1} + \frac{D_2}{2\delta_2} + \frac{\lambda \Delta z}{2\delta_1 \delta_2}$
1. $(1+\frac{\Delta z}{R}\delta_1 - \frac{\lambda \Delta z}{D_1} \le \delta_2 \le (1+\frac{\Delta z}{R})\delta_1 + \frac{\lambda \Delta z}{D_1}$
1. $N \ge \frac{\lambda \Delta z}{\delta_1 \delta_2}$

```python
import numpy as np
import matplotlib.pyplot as plt

D1 = 2e-3
D2 = 4e-3
Dz = 0.1
wvl = 1e-6
d1 = np.arange(0,30e-6,0.1e-6)
Nsamp = 2**np.arange(7,13)

fig,ax = plt.subplots()
ax.grid(True)
ax.set_xlim(0,30e-6)
ax.set_ylim(0,50e-6)
ax.ticklabel_format(axis='both',style='sci',scilimits=(-6,-6))

#### Constraint 1 ####
d2_c1 = ((-D2/D1)*d1) + (wvl*Dz)/D1
ax.plot(d1,d2_c1)

for N in Nsamp:
    d2_c2 = ((D2/2)+((wvl*Dz)/(2*d1)))\
        /(N-(D1/(2*d1)))
    ax.plot(d1,d2_c2)

plt.show()
```
