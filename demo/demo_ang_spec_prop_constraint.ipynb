# Fresnel diffraction integral sampling constraints

There are 4 inequalities that govern the sampling required for Fresnel diffraction.

1. $\delta_2 \le -\frac{D_2}{D_1}\delta_1 + \frac{\lambda \Delta z}{D_1}$
1. $N \ge \frac{D_1}{2\delta_1} + \frac{D_2}{2\delta_2} + \frac{\lambda \Delta z}{2\delta_1 \delta_2}$
1. $(1+\frac{\Delta z}{R}\delta_1 - \frac{\lambda \Delta z}{D_1} \le \delta_2 \le (1+\frac{\Delta z}{R})\delta_1 + \frac{\lambda \Delta z}{D_1}$
1. $N \ge \frac{\lambda \Delta z}{\delta_1 \delta_2}$

```python
"""
Demo of the angular spectrum propagation constraints 4, 3, and 1.
"""
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm,ticker

D1 = 2e-3
D2 = 4e-3
Dz = 0.1
wvl = 1e-6
d1 = np.arange(0,30e-6,0.5e-6)
d2 = np.arange(0,60e-6,0.5e-6)
R = 1

fig,ax = plt.subplots()

#### Constraint 1 ####
d2_c1 = ((-D2/D1)*d1) + (wvl*Dz)/D1
ax.plot(d1,d2_c1,linestyle='dashdot',color='green')

#### Constraint 3 ####
d2_c3 = ((1+(Dz/R))*d1) + ((wvl*Dz)/(D1))   # Constraint 3 Upper Bound
ax.plot(d1,d2_c3,linestyle='dashed',color='purple')

#### Constraint 4 ####
delta1,delta2 = np.meshgrid(d1,d2)
N = (wvl*Dz)/(delta1*delta2)
Nlabels = 2**np.arange(5,20)
CS = ax.contourf(d1,d2,N,levels=Nlabels,locator=ticker.LogLocator(base=2.0), cmap=cm.PuBu_r)
fig.colorbar(CS)

ax.grid(True)
ax.clabel(CS,fontsize=9,colors='black')
ax.ticklabel_format(axis='both',style='sci',scilimits=(-6,-6))
ax.set_title("Angular-spectrum propagation - Constraints 4,3 and 1")
ax.set_xlabel(r"$\delta_1[\mu m]$",math_fontfamily='cm')
ax.set_ylabel(r"$\delta_2[\mu m]$",math_fontfamily='cm')
ax.grid(True)
ax.set_xlim(0,30e-6)
ax.set_ylim(0,60e-6)
ax.ticklabel_format(axis='both',style='sci',scilimits=(-6,-6))

plt.show()
```
