```python
import numpy as np
import matplotlib.pyplot as plt
from optprop.grid import square_meshgrid
from optprop.transforms import cart2pol
from optprop.aperture_functions import point_source

D = 8e-3
wvl = 1e-6
k = (2*np.pi)/wvl
Dz = 1
arg = D/(wvl*Dz)
delta1 = 1/(10*arg)
delta2 = D/100
N = 1024
x1,y1 = square_meshgrid(N,delta1)
A = wvl * Dz
pt = point_source(x1,y1,A,D,wvl,Dz)

fig,ax = plt.subplots()
ax.plot(x1[0,:],np.abs(pt[N//2,:])**2,marker='.')
ax.set_xlabel(r"$x[m]$",math_fontfamily='cm')
ax.set_ylabel(r"Irradiance $[W/m^2]$",math_fontfamily='cm')
ax.ticklabel_format(axis='y',style='sci',scilimits=(3,3))
ax.ticklabel_format(axis='x',style='sci',scilimits=(-3,-3))
ax.set_xlim(-1e-3,1e-3)
ax.set_ylim(0,5e3)
ax.grid(True,'both')
plt.show()
```
