Demo of rect() function in 1D

```python
import numpy as np
import matplotlib.pyplot as plt
from optical_functions.aperture_functions import rect

#### Configs ####
DISPLAY_FIGS = True

#### Optical parameters ####
L = 2           # Spacial extent of grid [m]
N = 20          # Number of sample points in 1D
delta = L/N     # Spacial interval [m]

#### Create function ####
x = np.arange(-N/2,N/2,1) * delta
y = rect(x)

#### Display figures ####
if DISPLAY_FIGS:
    fig,ax = plt.subplots()
    ax.plot(x,y,marker='x')
    ax.set_xlabel(r"$x [m]$",math_fontfamily="cm")
    ax.set_ylabel(r"$y(x)$",math_fontfamily="cm")
    ax.set_title("Rectangular Function")
    ax.grid(visible=True,which="major")
    plt.show()
```

Demo of circ() function in 2D

```python
import numpy as np
import matplotlib.pyplot as plt
from optical_functions.aperture_functions import circ

#### Configs ####
DISPLAY_FIGS = True

#### Optical parameters ####
L = 5           # Spacial extent of grid [m]
N = 20          # Number of sample points in 1D
D = 2           # Circle diameter [m]
delta = L/N     # Spacial interval [m]

#### Create function ####
x = np.arange(-N/2,N/2,1) * delta
y = x
x_mesh,y_mesh = np.meshgrid(x,y)
z = circ(x_mesh,y_mesh,D)

#### Display figures ####
if DISPLAY_FIGS:
    fig,ax = plt.subplots()
    im1 = ax.imshow(z,extent=[-L/2,L/2,-L/2,L/2])
    fig.colorbar(im1)
    ax.set_xlabel(r"$x [m]$",math_fontfamily="cm")
    ax.set_ylabel(r"$y [m]$",math_fontfamily="cm")
    ax.set_title("Circular Function")
    ax.grid(visible=True,which="both")
    plt.show()
```
