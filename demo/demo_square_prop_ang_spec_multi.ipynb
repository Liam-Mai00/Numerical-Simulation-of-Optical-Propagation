```python
import numpy as np
import matplotlib.pyplot as plt
from optprop.grid import square_meshgrid
from optprop.aperture_functions import rect
from optprop.prop import ang_spec_multi_prop_vac

D1 = 2e-3
D2 = 6e-3
wvl = 1e-6
k = 2*np.pi/wvl
z = 2
delta1 = D1/30
deltan = D2/30
N = 128
n = 5
z = np.arange(1,n+1) * z/n
x1,y1 = square_meshgrid(N,delta1)
ap = rect(x1/D1) * rect(y1/D1)
Uout,x2,y2 = ang_spec_multi_prop_vac(ap,wvl,delta1,deltan,z)

fig,ax = plt.subplots(1,2)
fig.set_figwidth(10)
ax[0].plot(x2[0,:],np.abs(Uout[N//2,:])**2,marker='x')
ax[0].set_xlim(-3e-3,3e-3)
ax[0].set_ylim(0,3)
ax[0].ticklabel_format(axis='x',style='sci',scilimits=(-3,-3))
ax[0].grid(True,'both')
ax[0].set_xlabel(r"$x_n[mm]$",math_fontfamily='cm')
ax[0].set_ylabel("Irradiance")
ax[0].set_title("Irradiance at observation plane")
ax[1].plot(x2[0,:],np.angle(Uout[N//2,:]),marker='x')
ax[1].set_xlabel(r"$x_n[mm]$",math_fontfamily='cm')
ax[1].set_ylabel("Phase[rad]")
ax[1].ticklabel_format(axis='x',style='sci',scilimits=(-3,-3))
ax[1].set_xlim(-3e-3,3e-3)
ax[1].set_ylim(-3,3)
ax[1].grid(True,'both')
ax[1].set_title("Phase at observation plane")
plt.show()
```
