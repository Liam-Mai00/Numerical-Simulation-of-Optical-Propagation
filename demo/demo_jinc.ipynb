# **Sombrero function**
Sombrero function (also known as jinc function) is based on the sinc function and is defined through the Bessel function of the first kind $J_1$.<br>
$$somb(\rho)=\frac{2J_1(\pi\rho)}{\pi\rho}$$
Where $\rho^2=x^2+y^2$ <br>
[See more](https://en.wikipedia.org/wiki/Sombrero_function) on the Wiki.

```python

%config InlineBackend.print_figure_kwargs = {'bbox_inches':None}
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from scipy.special import jv

def somb(rho):
    y = (2*jv(1,np.pi*rho))/(np.pi*rho)
    return y

x = np.arange(-2,2,0.1)
y = x
x,y = np.meshgrid(x,y)
rho = x**2 + y**2

somb_out = somb(rho)

fig1 = plt.figure()
fig1.set_figwidth(20)
ax1 = fig1.add_subplot(111,projection='3d')
ax1.plot_surface(x,y,somb_out)
ax1.set_title("Surface plot of the jinc function")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_zlabel(r"$somb(\rho)$",math_fontfamily='cm')

fig2,ax2 = plt.subplots()
im1 = ax2.contourf(x,y,somb_out)
ax2.set_title("Contour plot of the jinc function")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
fig2.colorbar(im1)
plt.show()
```
