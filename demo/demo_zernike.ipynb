# Zernike Circular Polynomial Demo


```python
from optprop.zernike import zernike
from optprop.transforms import cart2pol, pol2cart
import numpy as np
import matplotlib.pyplot as plt

N = 128
L = 2
delta = L/N

x = np.arange(-N/2,N/2,1) * delta
x,y = np.meshgrid(x,x)  # Assume a square grid
r,theta = cart2pol(x,y)

modes = [6, 12, 24]
titles = ["X Primary Astigmatism", "X Secondary Astigmatism", "X Tertiary Astigmatism"]
for idx,i in enumerate(modes):
    Z = zernike(i,r,theta)
    fig,ax = plt.subplots()
    im = ax.imshow(Z,extent=[-L/2,L/2,-L/2,L/2])
    fig.colorbar(im)
    ax.set_xlabel(r"$x[m]$",math_fontfamily='cm')
    ax.set_ylabel(r"$y[m]$",math_fontfamily='cm')
    ax.set_title(f"{titles[idx]} ($i={i}$)",math_fontfamily='cm')
plt.show()
```
