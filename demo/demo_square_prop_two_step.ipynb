```python
import numpy as np
import matplotlib.pyplot as plt
from optprop.aperture_functions import rect
from optprop.prop import two_step_prop

N = 1024
L = 1e-2
delta1 = L/N
D = 2e-3
wvl = 1e-6
k = (2*np.pi)/wvl
Dz = 1
delta2 = (wvl*Dz)/(N*delta1)

x1 = np.arange(-N/2,N/2,1) * delta1
x1,y1 = np.meshgrid(x1,x1)
ap = rect(x1/D) * rect(y1/D)
Uout,x2,y2 = two_step_prop(ap,wvl,delta1,delta2,Dz)

fig,ax = plt.subplots(1,2)
ax[0].plot(x2[0,:],np.abs(Uout[N//2,:])**2,marker='x')
ax[0].ticklabel_format(axis='x',style='sci',scilimits=(-3,-3))
ax[0].set_xlim(-2.3e-3,2.5e-3)
ax[0].set_ylim(0,2.5)
ax[0].set_xlabel(r"$x[m]$",math_fontfamily='cm')
ax[0].set_ylabel("Irradiance")
ax[0].grid(True,'both')

ax[1].plot(x2[0,:],np.angle(Uout[N//2,:]),marker='x')
ax[1].ticklabel_format(axis='x',style='sci',scilimits=(-3,-3))
ax[1].set_xlim(-2.5e-3,2.5e-3)
ax[1].set_ylim(-3.1,3.1)
ax[1].set_xlabel(r"$x[m]$",math_fontfamily='cm')
ax[1].set_ylabel("Phase[rad]")
ax[1].grid(True,'both')

plt.show()
```
