## Understanding the literature
In chapter 5, section 5.2.1 Coherent Imaging, there is an equation defining the amplitude spread function $h(u,v)$. To reiterate for convenience:

$$
    h(u,v) = \frac{1}{\lambda z_i}\mathcal{F}\{\mathcal{P}(x,y)\}_{f_x=\frac{u}{\lambda z_i},f_y=\frac{v}{\lambda z_i}}
$$
The spacial sample spacings are hinted next to the fourier transform operation $\mathcal{F}\{\cdot\}$. To be explicit, the hints are: $f_x=\frac{u}{\lambda z_i}$ and $f_y=\frac{v}{\lambda z_i}$. Using proper terminology, the equation is *evaluated* with these conditions.<br>

From previous chapters, the spacial frequencies $(f_x,f_y)$ are defined as $f_x = \frac{1}{N\delta_1}$. Assuming a square grid, naturally, $f_y$ is the same. The 1D coordinates of frequencies can then be coded as:

```python
import numpy as np

# Number of samples
N = 64
# Spacial extent of source plane [m]
L = 1
# Spacial interval of source plane [m]
delta = L/N
# Spacial frequency interval [1/m]
delta_f = 1/(N*delta)

# Define 1D spacial coordinates
x = np.arange(-N/2,N/2,1) * delta
print(x)
# Define 1D frequency coordinates
fX = np.arange(-N/2,N/2,1) * delta_f
print(fX)
```

From earlier, the newly defined coordinates $(u,v)$ can be deduced from the evaluation conditions. Rearranging the conditions gives us the new definitions:
$$
    f_x = \frac{u}{\lambda z_i}
$$
$$
    f_x\lambda z_i = u
$$
And since $f_x = \frac{1}{N\delta_1}$ The new coordinate system $(u,v)$ has spacing of:
$$
u = \frac{\lambda z_i}{N\delta_1}
$$
And this can be coded into:


```python
import numpy as np

N = 64
L = 0.1
delta = L/N
wvl = 1e-6
z = 0.25
U = (wvl*z)/(N*delta)
u = np.arange(-N/2,N/2,1) * U
print(u)
```

An additional note, a very obvious analysis is that the value associated with index 1 is the sample spacing. For example, given `N = 5` then `idx = np.arange(-N/2,N/2,1)` would yield `[-5, -4, -3, -2, -1, 0, 1, 2, 3, 4]` after multiplying this array by $\delta_1$ then the index 1 **is** the sample spacing.
