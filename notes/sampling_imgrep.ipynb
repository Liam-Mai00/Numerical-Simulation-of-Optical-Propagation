# Sampling coordinates vs Image representation
There is an important distinction between the sampling coordinates and the image representation. This is why a circle can look 'off centre' when displayed as an image with something like `imshow`
## Sampling coordinates
Let's say we have a physical grid (spacial), that is 2m x 2m `L = 2`. We can create a system on how we go about sampling this grid. Given sample number `N = 4`, sampling at `x = [0.0, 0.5, 1.0, 1.5]`m is not conventional at all, so we use the cartesian system with the origin at the centre which is commonly used in optics. (Notice that 2.0 is missing? This is because we can't break the `N = 4` rule we mentioned earlier otherwise we'd end up with 5 values so we cannot sample any further) Therefore, `x = [-1.0, -0.5, 0.0, 0.5]` makes sense. Assuming a 2D square grid, `y` will have the same values <br>
To link all the aforementioned variables, it can be written in code as:

```python
import numpy as np

L = 2               # Spacial extent of grid [m]
N = 4               # Number of sample points
delta = L/N         # Sample interval [m]
x = np.arange(-N/2,N/2,1) * delta 
print(x)
```

Given this coordinate system, we can now measure the field intensity's **point** value at (-1.0,-1.0) then do the same for (-1.0,-0.5) and so on until (0.5,0.5).
## Image representation
The field intensity at (-1.0,-1.0) is associated with a value e.g. `1+1j`. However, we must now represent this as an image for plotting. Sampling the space gives us `N*N` values. In this case, it's 16 values. This means that we will see 16 **pixels**. Each pixel area will represent **one** field intensity value. Therefore, it is assumed that the value is now located at the centre of the pixel but in reality, the entire pixel represents that value.
