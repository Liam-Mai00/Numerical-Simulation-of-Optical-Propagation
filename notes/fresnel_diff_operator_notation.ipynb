# Fresnel Diffraction in Vacuum

When light propagates at a sufficiently close distance, it propagates in the **near-field**. Additionally, it no longer propagates under the Fraunhofer regime, but the Fresel one. The calculation in the near field is more sophisticated compared to the far field.<br>

## Different forms of Fresnel Diffraction Integral

The **Fresnel Diffraction Integral** is defined as:

\begin{equation}
    U(x_2,y_2) = \frac{e^{ik\Delta z}}{i\lambda \Delta z}\int_{-\infty}^{\infty} \int_{-\infty}^{\infty} U(x_1,y_1)e^{i\frac{k}{2\Delta z}[(x_2 - x_1)^2+(y_2-y_1)^2]}dx_1dy_1
\end{equation}

There are 2 ways of computing the Fresnel Diffraction Integral:
+ Expanding the squared terms in the exponential and factoring portions out of the integral:

\begin{equation}
    U(x_2,y_2) = \frac{e^{ik\Delta z}}{i\lambda \Delta z}e^{i\frac{k}{2\Delta z}(x_2^2+y_2^2)}\int_{-\infty}^{\infty}\int_{-\infty}^{\infty}U(x_1,y_1)e^{i\frac{k}{2\Delta z}(x_1^2+y_1^2)}e^{-i\frac{2\pi}{\lambda \Delta z}(x_2x_1+y_2y_1)}dx_1dy_1
    \tag{2}
\end{equation}

The above equation can be computed as a single FT.

+ Second form of the diffraction integral can be computed as a convolution of the source plane field $U(x_1,y_1)$ and the free-space amplitude spread function:

\begin{equation}
    U(x_2,y_2) = U(x_1,y_1) \otimes [\frac{e^{ik\Delta z}}{i\lambda \Delta z}e^{i\frac{k}{2\Delta z}(x_1^2+y_1^2)}]
    \tag{3}
\end{equation}

This can be computed via the product of 2 FTs. 

## Operator Notation

The following table contains symbols for Fresnel Diffraction in a vacuum and operator notation described by Nazarathy and Shamir.

Symbol | Meaning |
| --- | --- |
$\bold{r}_1 = (x_1,y_1)$ | Source-plane coordinates |
$\bold{r}_2 = (x_2,y_2)$ | Observation-plane coordinates |
$\delta_1$ | Grid spacing in source plane
$\delta_2$ | Grid spacing in observation plane
$\bold{f}_1 = (f_{x_1},f_{y_1})$ | Spatial frequency of source plane |
$\delta_{f_1}$| Grid spacing in source-plane spatial frequency |
$z_1$ | Location of source plane along the optical axis |
$z_2$ | Location of observation plane along the optical axis |
$\Delta z$ | Distance between the source plane and observation plane |
$m$ | Scaling factor from source plane to observation plane |

The operator notation described by Nazarathy and Shamir contains several important identities:

\begin{equation}
    \mathcal{Q}[c,\bold{r}]\{U(\bold{r})\} \equiv e^{i\frac{k}{2}c|\bold{r}|^2}U(\bold{r})
    \tag{4.1}
\end{equation}

\begin{equation}
    \mathcal{V}[b,\bold{r}]\{U(\bold{r})\} \equiv bU(b\bold{r})
    \tag{4.2}
\end{equation}

\begin{equation}
    \mathcal{F}[\bold{r},\bold{f}] \equiv \int_{-\infty}^\infty U(\bold{r})e^{-i2\pi \bold{f}\cdot \bold{r}}d\bold{r}
    \tag{4.3}
\end{equation}

\begin{equation}
    \mathcal{F}^{-1}[\bold{f},\bold{r}] \equiv \int_{-\infty}^\infty U(\bold{f})e^{i2\pi \bold{f}\cdot \bold{r}}d\bold{f}
    \tag{4.4}
\end{equation}

\begin{equation}
    \mathcal{R}[d,\bold{r}_1,\bold{r}_2]\{U(\bold{r}_1\} \equiv \frac{1}{i\lambda d} \int_{-\infty}^\infty U(\bold{r}_1)e^{i\frac{k}{2d}|\bold{r}_2-\bold{r}_1|^2}d\bold{r}_1
    \tag{4.5}
\end{equation}

J. Schmidt has defined an additional notation for convenience later on but is not mentioned in Nazarathy and Shamir's paper:

\begin{equation}
    \mathcal{Q}_2[d,\bold{r}]\{U(\bold{r}\} \equiv e^{i\pi^2\frac{2d}{k}|\bold{r}|^2}U(\bold{r})
    \tag{4.6}
\end{equation}

However, it can also be represented with Nazarathy and Shamir notation as:
$$\mathcal{Q}_2[d,\bold{r}] = \mathcal{Q}[\frac{4\pi^2}{k^2}d,\bold{r}]$$

### Understanding the notation
- The arguments of the notations are placed in square brackets [...]
- Operands are located within the curly brackets {...}
- Domain of operand is the *first* parameter
- Domain of result is the *second* parameter

**Example:**

$$
\mathcal{F}[\bold{r,f}]\{U(\bold{r})\} \equiv \int_{-\infty}^{\infty}U(\bold{r})e^{-i2\pi\bold{f\cdot r}}d\bold{r}
$$

In here, $\bold{r}$ and $\bold{f}$ are the operator parameters. Since this is the Fourier Transform operation $\mathcal{F\{\cdot\}}$, we know that this operation transforms from the spacial domain to a frequency one. Therefore, $\bold{r}$ is the spacial coordinate system and is placed first since it is the operand's domain. Similarly, $\bold{f}$ is the frequency coordinate system and is placed second as that will be the operation output's domain.<br>
As for the operand itself (the thing that gets operated on), it is the input field $U(\bold{r})$ because it's located within the curly brackets. Finally, the expression located on the right is simply a mathematical representation of the operator notation. For additional notes, check out [Nazarathy and Shamir's work](https://opg.optica.org/josa/abstract.cfm?uri=josa-70-2-150).<br>
It is helpful to note that when combining the operator notations, you typically want to read from **right to left**. In the example of the one-step propagation: <br>

$$
U(\bold{r}_2) = \mathcal{R}[\Delta z,\bold{r}_1,\bold{r}_2]\{U(\bold{r}_1)\} = \mathcal{Q}\left[\frac{1}{\Delta z},\bold{r}_2\right]\mathcal{V}\left[\frac{1}{\lambda \Delta z},\bold{r}_2\right]\mathcal{F}\left[\bold{r}_1,\bold{f}_1\right]\mathcal{Q}\left[\frac{1}{\Delta z},\bold{r}_1\right]\{U(\bold{r}_1)\}
$$
- $\mathcal{Q}$ operates on $U(\bold{r}_1)$
- $\mathcal{F}$ operates on $\mathcal{Q}$ and $U(\bold{r}_1)$
- $\mathcal{V}$ operates on $\mathcal{F}$, $\mathcal{Q}$, and $U(\bold{r}_1)$
- $\mathcal{Q}$ operates on $\mathcal{V}$, $\mathcal{F}$, $\mathcal{Q}$, and $U(\bold{r}_1)$
