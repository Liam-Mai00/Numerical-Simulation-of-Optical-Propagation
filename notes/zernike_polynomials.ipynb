# Zernike Polynomials

Zernike polynomials have different variants, such as:

- Zernike circle polynomials (Orthogonal over a circular aperture)
- Zernike annular polynomials (Orthogonal over an annular aperture)
- Zernike-Gauss circular polynomials (Orthogonal over a gaussian aperture)
- Zernike-Gauss annular polynomials (Orthogonal over a gaussian & annular aperture)

We will work with circular apertures as they are most common therefore Zernike circle polynomials are deployed.<br>

The Zernike circle polynomials are defined as:
\begin{equation}
    Z_n^m(r,\theta) = \sqrt{2(n+1)}R_n^m(r)G^m(\theta)
\end{equation}

Where $m,n \in \Bbb{Z^{0+}}$ (means $m$ and $n$ are non-negative integers - *This includes zero*) and $m\le n$. Writing $Z_n^m(r,\theta)$ as a single index $Z_i(r,\theta)$ for simplicity gives:<br>
<font color='red'>[This set notation could be incorrect]</font>

\begin{equation}
Z_i(r,\theta) = \begin{cases}
    \sqrt{2(n+1)}R_n^m(r)G^m(\theta) & m \neq 0 \\ % & is your "\tab"-like command (it's a tab alignment character)
    \sqrt{n+1}R_n^0(r) & m = 0
    \tag{2}
\end{cases}
\end{equation}

Mapping from $(n,m) \to i$ is complicated. Radial and azimuthal factors $R_n^m(r)$ and $G^m(\theta)$ are given as:

\begin{equation}
    R_n^m(r) = \sum_{s=0}^{(n-m)/2}\frac{(-1)^s(n-s)!}{s!(\frac{n+m}{2}-s)!(\frac{n-m}{2}-s)!}r^{n-2s}
    \tag{3.1}
\end{equation}

\begin{equation}
G^m(\theta) =
\begin{cases}
    sin(m\theta) & i \text{ odd}\\
    cos(m\theta) & i \text{ even}
\end{cases}
\tag{3.2}
\end{equation}

$i$ is the mode number and normalised polar coordinates on the unit circle. Note that $s! = \Gamma (s+1)$ within the code because gamma functions are computed faster than factorials. The mapping from $(n,m) \to i$ is provided in the `zernike_index.csv` file.
