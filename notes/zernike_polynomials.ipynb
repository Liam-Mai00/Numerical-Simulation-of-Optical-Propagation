# Zernike Polynomials

Zernike polynomials have different variants, such as:

- Zernike circle polynomials (Orthogonal over a circular aperture)
- Zernike annular polynomials (Orthogonal over an annular aperture)
- Zernike-Gauss circular polynomials (Orthogonal over a gaussian aperture)
- Zernike-Gauss annular polynomials (Orthogonal over a gaussian & annular aperture)

We will work with circular apertures as they are most common therefore Zernike circle polynomials are deployed.<br>

The Zernike circle polynomials are defined as:
\begin{equation}
    Z_n^m(r,\theta) = \sqrt{2(n+1)}R_n^m(r)G^m(\theta)
\end{equation}
Where $m,n \in \Bbb{Z^{0+}}$ (means $m$ and $n$ are non-negative integers - *This includes zero*) <font color='red'>[This set notation could be incorrect]</font>
