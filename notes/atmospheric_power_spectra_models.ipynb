# Common models for atmospheric power spectra

## Kolmogorov refractive-index power spectral density

\begin{equation}
    \Phi_n^K(\kappa) = 0.033C_n^2\kappa^{-11/3}
\end{equation}

## Von Karman refractive-index power spectral density

\begin{equation}
    \Phi_n^{vK}(\kappa) = \frac{0.033C_n^2}{(\kappa^2 + \kappa_0^2)^{11/6}}
    \tag{2}
\end{equation}

## Modified Von Karman refractive-index power spectral density

\begin{equation}
    \Phi_n^{mvK}(\kappa) = 0.033C_n^2\frac{exp(-\kappa^2/\kappa_m^2)}{(\kappa^2+\kappa_0^2)^{11/6}}
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

#### PSD Properties ####
Cn2 = 10**-13       # Refractive-index structure parameter [m^-2/3]
l0 = 10**-4         # Inner scale [m]
L0 = 10**2          # Outer scale [m]          
k0 = 2*np.pi/L0
km = 5.92/l0

#### Display PSDs ####
k = np.linspace(1/L0,1/l0,1000)
PSD_K = 0.033*Cn2*k**(-11/3)
PSD_vK = (0.033*Cn2)/((k**2 + k0**2)**(11/6))
PSD_mvK = 0.033*Cn2*(np.exp(-k**2/km**2))/((k**2+k0**2)**(11/6))

fig,ax = plt.subplots()
ax.loglog(k,PSD_K/(0.033*Cn2))
ax.loglog(k,PSD_vK/(0.033*Cn2))
ax.loglog(k,PSD_mvK/(0.033*Cn2))
ax.set_xlabel(r"Spacial Frequency $\k\left[rad/m\right]$",math_fontfamily='cm')
ax.set_ylabel(r"$\Phi_n(\kappa)/\left[0.033C_n^2\right]$",math_fontfamily='cm')
ax.set_xlim(10**-2,10**4)
ax.set_ylim(10**-20,10**10)
ax.grid(True,'both')
```
