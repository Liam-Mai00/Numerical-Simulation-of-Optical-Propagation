# Angular-Spectrum Propagation

Previously, the one-step and two-step propagation techniques leverages a single Fourier Transform operation. Where, the two-step propagation introduces a more flexible way of choosing the observation-plane spacing $\delta_2$. The intermidate propagation allows us to choose $z_{1a}$ such that $m$ (effectively $\delta_2$) can be at a desired value.<br>
The Angular-Spectrum Propagation evaluates the Fresnel diffraction integral as a **convolution**. The angular spectrum evaluation is given as:
\begin{equation}
U(\bold{r}_2) = \mathcal{F}^{-1}[\bold{f}_1,\bold{r}_2]H(\bold{f}_1)\mathcal{F}[\bold{r}_1,\bold{f}_1]\{U(\bold{r}_1\}
\end{equation}
Where $H(\bold{f}_1)=e^{ik\Delta z}e^{-i\pi \lambda \Delta z(f_{x_1}^2+f_{y_1}^2)}$ and is the transfer function of free space propagation.<br>
