
# **Bessel function of first kind**
Bessel functions are special functions commonly appearing in wave motion, heat conduction, and other phenomena. They are solutions to a type of ordinary differetial equation:
$$x^2\frac{d^2y}{dx^2} + x\frac{dy}{dx} + (x^2 - \alpha^2)y = 0$$
Where: <br>
$\alpha$ = Bessel function order <br>

[See more](https://en.wikipedia.org/wiki/Bessel_function) on Bessel functions from the Wiki

## **Bessel function demo**
The function is used within jinc() from the aperture_functions module.<br>
Applying Frobenius method on Bessel's equation:
$$J_\alpha(x) = \sum_{m=0}^\infty \frac{(-1)^m}{m!\Gamma(m+\alpha+1)} (\frac{x}{2})^{2m+\alpha}$$
<br>

The parameters are based on the [Wiki image](https://en.wikipedia.org/wiki/Bessel_function#/media/File:BesselJ.png)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jv

DISPLAY_FIGS = True

# Set Bessel function parameters
x = np.arange(0,22,0.1)
alpha = np.array([0,1,2])

# Plot Bessel function with each order specified
if DISPLAY_FIGS:
    fig,ax = plt.subplots()
    for order in alpha:
        y = jv(order,x)
        ax.plot(x,y,label=rf"$J_{order}(x)$")
    ax.set_xlabel(r"$x$",math_fontfamily='cm')
    ax.set_ylabel(r"$J_\alpha(x)$",math_fontfamily='cm')
    ax.set_title(rf"Bessel function of the first kind $\alpha\in{alpha.tolist()}$",math_fontfamily='cm')
    ax.legend()
    ax.grid(visible=True,which='both')
    plt.show()
```
